```python
import requests
import json

BASE_URL = "http://localhost:8000"

def pretty(response):
    """Print formatted JSON response."""
    print(f"Status: {response.status_code}")
    print(json.dumps(response.json(), indent=2))
```

```python
resp = requests.get(f"{BASE_URL}/health")
pretty(resp)
```

```python
resp = requests.get(f"{BASE_URL}/analyze/RELIANCE.NS")
pretty(resp)
```

```python
resp = requests.get(f"{BASE_URL}/explain/TCS.NS")
pretty(resp)
```

```python
resp = requests.post(
    f"{BASE_URL}/ask",
    json={"question": "What does a Bollinger Band squeeze mean?",
          "ticker": "RELIANCE.NS"}
)
pretty(resp)
```

```python
resp = requests.get(f"{BASE_URL}/overview")
pretty(resp)
```
