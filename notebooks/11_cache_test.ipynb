```python
import requests
import time
```

```python
BASE_URL = "http://localhost:8000"
```

# First request — cache miss (slow)

```python
start = time.time()
resp1 = requests.get(f"{BASE_URL}/analyze/RELIANCE.NS")
time1 = time.time() - start
print(f"First request  : {time1:.2f}s (cache miss)")
```

# Second request — cache hit (fast)

```python
start = time.time()
resp2 = requests.get(f"{BASE_URL}/analyze/RELIANCE.NS")
time2 = time.time() - start
print(f"Second request : {time2:.2f}s (cache hit)")
print(f"Speedup        : {time1/time2:.0f}x faster")
```

# Verify same result

```python
assert resp1.json()["probability_up"] == resp2.json()["probability_up"]
print("Results identical ✓")
```
